```python
import os
# date and time manipulation
import datetime as dt
import pytz
tz = pytz.timezone("Etc/GMT")

# data manipulation
import numpy as np
import pandas as pd


# sentiment analysis modules
import preprocessor as p # filters the unwanted elements of the tweets
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer # sentiment analysis module
SIA = SentimentIntensityAnalyzer()


```

```python
with open(os.path.join(os.getcwd(),"tweets.csv")) as fp:
    raw_tweets_data = pd.read_csv(fp)
raw_tweets_data['time'] = pd.to_datetime(raw_tweets_data['date'],errors='coerce')
raw_tweets_data.dropna(inplace=True)
```

```python
raw_tweets_data.drop('date',axis=1,inplace=True)
```

```python
raw_tweets_data
```

```python
raw_tweets_data[['neg','neu','pos','comp']] = raw_tweets_data['content'].apply(lambda x: np.fromiter(SIA.polarity_scores(p.clean(x)).values(), dtype=float))
```
